# Quartiles and Boxplot

Quartiles and boxplots are valuable tools in descriptive statistics. Quartiles divide a dataset into four equal parts, providing insight into the data's distribution. A boxplot visually represents this information, using quartiles to display the spread, median, and potential outliers. These tools are used to summarize data distribution, identify outliers, compare distributions across different groups, and understand the variability within a dataset, particularly through the interquartile range which highlights the spread of the central 50% of the data.

## Quartiles

Quartiles are specific percentiles that divide a dataset into four equal parts: the first quartile (Q1) is the 25th percentile, the second quartile (Q2) is the 50th percentile (which is also the median), and the third quartile (Q3) is the 75th percentile. Percentiles, in general, indicate the value below which a given percentage of observations in a dataset falls.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/ProgWdata/refs/heads/main/grades_ok.csv')
df.head(3)
```

```python
print(df["Grade"].describe())
```

## Boxplot

A boxplot is a graphical representation of data distribution that shows quartiles, median, and potential outliers. The box represents the interquartile range (Q1 to Q3), the line inside is the median (Q2), and the whiskers extend to show the data range, excluding outliers.

```python
import seaborn as sns
sns.boxplot(df["Grade"]).set(**{})
```

```python
sns.boxplot(data=df,y="Grade")
```

```python
sns.boxplot(data=df,y="Grade",x="Course")
```

```python
sns.boxplot(data=df,y="Grade",x="Course",hue='Score')
```
