```python

%%writefile dash.py
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Auto Insurance Explorer", layout="wide")

st.title("🚗 Auto Insurance Data Explorer")

# CSV Source
CSV_URL = "https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/autoinsurance.csv"

# Load data
@st.cache_data
def load_data(url):
    return pd.read_csv(url)

df = load_data(CSV_URL)

# -------------------------
# Sidebar controls
# -------------------------
st.sidebar.header("Filters")

# State filter
states = st.sidebar.multiselect(
    "Select States",
    df["us_state"].unique(),
    placeholder="Choose one or more..."
)

filtered_df = df[df["us_state"].isin(states)] if states else df.copy()

# Year selector
year_cols = [col for col in df.columns if col.startswith("Y")]
year = st.sidebar.selectbox("Select Year", year_cols)

# Range filter
min_val = int(filtered_df[year].min())
max_val = int(filtered_df[year].max())

value_range = st.sidebar.slider(
    f"Value Range for {year}",
    min_val, max_val, (min_val, max_val)
)

filtered_df = filtered_df[
    (filtered_df[year] >= value_range[0]) &
    (filtered_df[year] <= value_range[1])
]

# Search box
query = st.sidebar.text_input("Search state")
if query:
    filtered_df = filtered_df[
        filtered_df["us_state"].str.contains(query, case=False)
    ]

# Normalization toggle
normalize = st.sidebar.toggle("Normalize Chart Values")

# -------------------------
# Main Tabs
# -------------------------
tab1, tab2, tab3 = st.tabs(["📄 Table", "📊 Stats", "📈 Charts"])

# ---- TABLE TAB ----
with tab1:
    st.subheader("Filtered Dataset")
    st.dataframe(filtered_df, use_container_width=True)

    st.download_button(
        "Download Filtered Data",
        filtered_df.to_csv(index=False).encode(),
        file_name="filtered_autoinsurance.csv",
        mime="text/csv"
    )

# ---- STATS TAB ----
with tab2:
    st.subheader("Summary Statistics")
    st.write(filtered_df.describe(include="all"))

# ---- CHART TAB ----
with tab3:
    st.subheader(f"Values for {year}")

    chart_data = filtered_df.set_index("us_state")[year]

    if normalize and len(chart_data) > 0:
        chart_data = (chart_data - chart_data.min()) / (chart_data.max() - chart_data.min())

    if len(chart_data) > 0:
        st.bar_chart(chart_data)
    else:
        st.warning("No data to display in chart. Adjust your filters.")
```
