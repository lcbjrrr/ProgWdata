```python

%%writefile dash.py


import streamlit as st
import pandas as pd


st.title("🚗 Auto Insurance Data Explorer")


df = pd.read_csv("https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/autoinsurance.csv")


st.sidebar.header("Filters")


states = st.sidebar.multiselect(
    "Select States",
    df["us_state"].unique(),
    placeholder="Choose one or more..."
)

filtered_df = df[df["us_state"].isin(states)] if states else df.copy()


year = st.sidebar.selectbox("Select Year", ["Y2021","Y2022","Y2023"])

if not filtered_df.empty:
    min_val = int(filtered_df[year].min())
    max_val = int(filtered_df[year].max())
    value_range = st.sidebar.slider(
            f"Value Range for {year}",
            min_val+1, max_val, (min_val+1, max_val)
        )
    filtered_df = filtered_df[
        (filtered_df[year] >= value_range[0]) &
        (filtered_df[year] <= value_range[1])
    ]
else:
    st.sidebar.warning("No data matches current filters.")



tab1, tab2, tab3 = st.tabs(["📄 Table", "📊 Stats", "📈 Charts"])

# ---- TABLE TAB ----
with tab1:
    st.subheader("Filtered Dataset")
    st.dataframe(filtered_df, use_container_width=True)

    if not filtered_df.empty:
        st.download_button(
            "Download Filtered Data",
            filtered_df.to_csv(index=False).encode(),
            file_name="filtered_autoinsurance.csv",
            mime="text/csv"
        )

# ---- STATS TAB ----
with tab2:
    st.subheader("Summary Statistics")
    if not filtered_df.empty:
        st.write(filtered_df.describe(include="all"))
    else:
        st.write("No data available for statistics.")

# ---- CHART TAB ----
with tab3:
    st.subheader(f"Values for {year}")
    if not filtered_df.empty:
        chart_data = filtered_df.set_index("us_state")[year]
        st.bar_chart(chart_data)
    else:
        st.warning("No data to display in chart. Adjust your filters.")
```
